```python
def  factorial_recur(n) :   #  ;
    if n == 1 :       
        return 1  
    else :
        return (n * factorial_recur(n - 1))     #    . 
    
    
def factorial_iter(n) :          # #  ;
    result = 1
    for k in range(1, n+1) :  # k:  1, 2, ..., n-1, n
        result = result * k    #  
    return result

print('Factorial(3) = ', factorial_recur(3))
print('Factorial(3) = ', factorial_iter(3))
```

```python
def insertion_sort(A) :
    n = len(A)   #  
    for i in range(1, n) :     #  : 1, 2, ... n-1
        key = A[i]          # A[i] key 
        j = i-1            #  A[i-1] 
        while j>=0 and A[j] > key :       #  
            A[j + 1] = A[j]           #     
            j = j - 1                   #   
        A[j + 1] = key                  #  A[i]   
        printStep(A, i)
def printStep(arr, val) :           #    
    print("  Step %2d = " % val, end='')    #   
    print(arr)        #   
data = [ 5, 3, 8, 4, 9, 1, 6, 2, 7 ]
print("Original  :", data)
insertion_sort(data)
print("Insertion :", data)
```
